### Problem 01 (15 points):

The data sets provided in the file "Class.csv" represents the number of points obtained by students at a quizz, where every row represents points obtained by a student.

Load the file and write a function which displays: 

* Number of students 

* Number of quizzes

* Average number of points for each student 

* Min number of points for each student 

* Max number of points for each student 

* Standard deviation of the number of points for each student 

* Average number of points for each quiz 

* Min number of points for each quiz 

* Max number of points for each quiz 

* Standard deviation of the number of points for each quiz 



````python
import numpy as np
def analyze_quiz_data(filename):
    # Load data
    data = np.loadtxt(filename, delimiter=',')

    
    # Calculate statistics for students (rows)
    student_means = np.mean(data, axis=1)
    student_mins = np.min(data, axis=1)
    student_maxs = np.max(data, axis=1)
    student_stds = np.std(data, axis=1)
    
    # Calculate statistics for quizzes (columns)
    quiz_means = np.mean(data, axis=0)
    quiz_mins = np.min(data, axis=0)
    quiz_maxs = np.max(data, axis=0)
    quiz_stds = np.std(data, axis=0)
    
    print("Quiz Analysis Results:")
    print(f"Number of students: {data.shape[0]}")
    print(f"Number of quizzes: {data.shape[1]}")
    print("\nStudent Statistics:")
    print(f"Average points per student: {student_means}")
    print(f"Min points per student: {student_mins}")
    print(f"Max points per student: {student_maxs}")
    print(f"Standard deviation per student: {student_stds}")
    print("\nQuiz Statistics:")
    print(f"Average points per quiz: {quiz_means}")
    print(f"Min points per quiz: {quiz_mins}")
    print(f"Max points per quiz: {quiz_maxs}")
    print(f"Standard deviation per quiz: {quiz_stds}")
analyze_quiz_data('Class.csv')
````

### Problem 02 (25 points)

2.1 (5 points) Create an array named $U$ of dimension 10 by 10, where each entry is drawn independently from the uniform distribution on $[0, 1]$

2.2 (5 points) Create an array named $G$ of dimension 10 by 10, where each entry is drawn independently from the normal (Gaussian) distribution with mean 0 and standard deviation 1


2.3 (15 points)

* Write a function which takes an array `A` as input, and returns two vectors:
    * Vector of averages over the rows of array `A`
    * Vector of averages over the columns of array `A`


* Write a function which takes an array `A` as input, and returns an array where each entry of `A` is squared


* Write a function which takes an array `A` as input, and returns an array where each entry of `A` is raised at the power k:


* Report the result of applying the functions developed in 2.1, 2.2, and 2.3 with `k=3`, on $U$, and then on $G$ 



````python
# 2.1 - Uniform distribution
U = np.random.uniform(0, 1, size=(10, 10))
    
# 2.2 - Normal distribution
G = np.random.normal(0, 1, size=(10, 10))

# 2.3 - Vectors of Array
def get_averages(A):
    # Returns row and column averages
    row_avgs = np.mean(A, axis=1)
    col_avgs = np.mean(A, axis=0)
    return row_avgs, col_avgs

def square_array(A):
    # Returns array with squared elements
    return np.power(A, 2)

def power_array(A, k):
    # Returns array with elements raised to power k
    return np.power(A, k)

# Results for U
print("\nResults for Uniform Distribution Array (U):")

print("\nOriginal U array:")
print(U)

row_avgs_U, col_avgs_U = get_averages(U)
print("\nRow averages of U:")
print(row_avgs_U)
print("\nColumn averages of U:")
print(col_avgs_U)

U_squared = square_array(U)
print("\nU squared:")
print(U_squared)

U_cubed = power_array(U, 3)
print("\nU cubed (k=3):")
print(U_cubed)

# Results for G
print("\nResults for Normal Distribution Array (G):")

print("\nOriginal G array:")
print(G)

row_avgs_G, col_avgs_G = get_averages(G)
print("\nRow averages of G:")
print(row_avgs_G)
print("\nColumn averages of G:")
print(col_avgs_G)

G_squared = square_array(G)
print("\nG squared:")
print(G_squared)

G_cubed = power_array(G, 3)
print("\nG cubed (k=3):")
print(G_cubed)
````

Sample four normal distributions, each with 1000 data points, respectively with means and standard deviation equal to:

* (m3, s3) = (-2, 0.5) 

* (m4, s4) = (1, 1) 

then plot each as a histogram with 50 bins on the same plot. Show a legend in the upper left corner of the plot with a custom label for each histogram (respectively `h1`, `h2`, `h3`, `h4`). Adjust the figure x-axis to go from $-7$ to $+7$, and the y-label to go from 0 to +80.


5.2 (15 points) 

Create a `Figure` object of width $10$ inches and height $8$ inches with $2 \times 2$ subplots (i.e. 2 subplots in each row and 2 in each column), and **plot each of the histograms above as a separate subplot** on this `Figure` object, using a tight layout.  For each sub-plot adjust the x-axis to go from $-7$ to $+7$, and the y-label to go from 0 to +80.



````python
#5.1 - normal distribution
data1 = np.random.normal(0, 2, 1000)
data2 = np.random.normal(2, 0.5, 1000)
data3 = np.random.normal(-2, 0.5, 1000)
data4 = np.random.normal(1, 1, 1000)
plt.hist(data1, bins=50, alpha=0.5, label='h1')
plt.hist(data2, bins=50, alpha=0.5, label='h2')
plt.hist(data3, bins=50, alpha=0.5, label='h3')
plt.hist(data4, bins=50, alpha=0.5, label='h4')
plt.xlim(-7, 7)
plt.ylim(0, 80)
plt.title('Histograms of Normal Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper left')

plt.show()

#5.2 - seperate subplots
Figure, axs = plt.subplots(2, 2, figsize=(10, 8), tight_layout=True)
axs[0, 0].hist(data1, bins=50, alpha=0.5)
axs[0, 0].set(xlim=[-7,7], ylim=[0,80], title='h1')

axs[0, 1].hist(data2, bins=50, alpha=0.5)
axs[0, 1].set(xlim=[-7,7], ylim=[0,80], title='h2')

axs[1, 0].hist(data3, bins=50, alpha=0.5)
axs[1, 0].set(xlim=[-7,7], ylim=[0,80], title='h3')

axs[1, 1].hist(data4, bins=50, alpha=0.5)
axs[1, 1].set(xlim=[-7,7], ylim=[0,80], title='h4')

for ax in axs.flat:
    ax.set(xlabel='Value', ylabel='Frequency')

plt.show()
````
